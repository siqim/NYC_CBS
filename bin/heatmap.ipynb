```python
import pandas as pd
import folium
import folium.plugins as plugins
import numpy as np
from folium.plugins import HeatMap
```

```python
plot_data = pd.read_csv('plot_data.csv')
```

```python
plot_data.head()
```

```python
posWithTime = []
time_index =[]
year = 2016
month = 9
second = 0
for day in [18, 19, 20, 21, 22, 23, 24]:
    for hour in range(0, 24):
        for minute in [0, 15, 30, 45]:
            timestamp = '%4d-%02d-%02d %02d:%02d:%02d'%(year, month, day, hour, minute, second)
            one_timestamp = plot_data[plot_data['trip_start_timestamp'] == timestamp][['pickup_centroid_latitude', 'pickup_centroid_longitude']]
            one_timestamp['counts'] = 1
            one_timestamp = one_timestamp.groupby(['pickup_centroid_latitude', 'pickup_centroid_longitude']).sum().reset_index()
            posWithTime.append(one_timestamp.values.tolist())
            time_index.append(timestamp)
            print(timestamp, len(one_timestamp))
```

```python
m = folium.Map([41.8973346, -87.6811209], tiles='openstreetmap', zoom_start=12)

hm = plugins.HeatMapWithTime(
    data=posWithTime,
    index=time_index,
    min_speed=1,
    max_speed=20,
    speed_step=1,
    radius=20,
    gradient={0.2: 'blue', 0.4: 'lime', 0.6: 'orange', 1: 'red'},
    min_opacity=0.5,
    max_opacity=0.8,
    use_local_extrema=True,
)

hm.add_to(m)

m.save('Omap.html')
```
